Dimension of the np.array

```python
# üëâ
#           The number of the [] from the () is the dimension(no of axis) in the numpy array.
# üëà

import numpy as np

# 0-D Array (Scalar)
a = np.array(10)
print(a.ndim)  # 0
# üëâ Type: 0-D Array (Scalar)
# Explanation: Only one single value ‚Äî no axes at all.
# It's just one element (like a single point).

# ----------------------------------------------------------------------------------

# 1-D Array (Vector)
b = np.array([1, 2, 3])
print(b.ndim)  # 1
# üëâ Type: 1-D Array (Vector)
# Explanation: Has 1 axis ‚Üí the horizontal line of elements.
# Axis 0 ‚Üí represents the position of each element [1, 2, 3].

# ----------------------------------------------------------------------------------

# 2-D Array (Matrix)
c = np.array([[1, 2], [3, 4]])
print(c.ndim)  # 2
# üëâ Type: 2-D Array (Matrix)
# Explanation: Has 2 axes ‚Üí rows and columns.
# Axis 0 ‚Üí rows, Axis 1 ‚Üí columns.
# So it's like:
# [ [1, 2],
#   [3, 4] ]

# ----------------------------------------------------------------------------------

# 3-D Array (Tensor)
d = np.array([
    [[1, 2], [3, 4]],
    [[5, 6], [7, 8]]
])
print(d.ndim)  # 3
# üëâ Type: 3-D Array (Tensor)
# Explanation: Has 3 axes.
# Axis 0 ‚Üí number of 2D matrices (or ‚Äúdepth‚Äù)
# Axis 1 ‚Üí rows within each matrix
# Axis 2 ‚Üí columns within each matrix
# Visually, it's like stacking two 2x2 matrices:
# Matrix 1:
# [[1, 2],
#  [3, 4]]
# Matrix 2:
# [[5, 6],
#  [7, 8]]
```

### Shape of the numpy array

```python
# The .shape attribute tells you how many elements are along each axis of the array and it‚Äôs like the blueprint of the array.
# shape return the (no_of_rows, no_of_columns)

import numpy as np

# 0-D array (scalar)
a = np.array(10)

# 1-D array (vector)
b = np.array([1, 2, 3, 4])

# 2-D array (matrix)
c = np.array([[1, 2, 3], [4, 5, 6]])

# 2d look like
#   [1,2,3]
#    4,5,6
# it has the 2 rows and coloums

# 3-D array (tensor)
d = np.array([
    [[1, 2], [3, 4]],
    [[5, 6], [7, 8]]
])

# Print type, dimension, and shape
arrays = [a, b, c, d]
names = ['0-D (Scalar)', '1-D (Vector)', '2-D (Matrix)', '3-D (Tensor)']

for name, arr in zip(names, arrays):
    print(f"{name}:")
    print(f"  ndim = {arr.ndim}")
    print(f"  shape = {arr.shape}\n")
```

### Joining / Adding 2 numpy arrays

```python
#To join 2 numpy arrays we make use of the concatenate() method

import numpy as np

n = np.array([1,2,3,4])
p = np.array([5,6,7,8])

resarray = np.concatenate((n,p))

print(resarray)

#Addition of arrays through Stack

# stack() -> Jaisa hai waisa hii merge the array
# vstack() --> vertically merge
# hstack() --> horizontal
# dstack() - 3D array (height wise)
# column_stack() - 2D array

# Using stack()

n1 = np.array([1,2,3,4,5])
n2 = np.array([5,6,7,8,9])

n3 = np.array([[1,2,3],[4,5,6]])
n4 = np.array([["a","b","c"],["d","e","f"]])

resultant_array = np.stack((n1,n2)) # the default axis is 0 which stacking row-wise
resultant2array = np.stack((n1,n2),axis = 0)
print("\n",resultant_array)
print("\n",resultant2array)
print(resultant_array.shape)
print(resultant2array.shape)

# Stacking using vstack()
resultantvarray = np.vstack((n1,n2))
print("\n\n", resultantvarray)
print(type(resultantvarray))

# Stacking using hstack()
resultantharray = np.hstack((n1,n2))
print("\n\n",resultantharray)


# Stacking of 2D arrays - return a 2D array itself.
# Stacking using dstack()
resultdarray = np.dstack((n3,n4))
print("\n\n",resultdarray)
print(resultdarray.shape)

# Stacking using column_stack()
resultcolumn_array = np.column_stack((n3,n4))
print("\n\n", resultcolumn_array)
print(resultcolumn_array.shape)

```

### Deleting Rows and Columns

```python
# drop() takes in two parameters- the name of the row/column you want to remove
# + the axis

# Syntax
# DataFrame.drop(name of columns, axis=0 or 1, inplace=False)



# Deleting columns using the drop() method.
import pandas as pd

data1 ={
    'Id': ['S01', 'S02', 'S03', 'S04', 'S05'],
    'Student': ['Steve', 'John', 'Jacob', 'David', 'Steve'],
    'Rank': [1, 4, 3, 5, 2],
    'Marks': [95, 70, 80, 60, 90]
}
dataframe1 = pd.DataFrame(data1, index = ["Index0","Index1","Index2","Index3","Index4"])
print(dataframe1)

resultframe = dataframe1.drop("Marks", axis = "columns")
print("\n\n",resultframe)

resultframe1 = dataframe1.drop("Marks", axis = 1)
print("\n\n",resultframe1)

# Deleting rows using drop() method
resultframe2 = dataframe1.drop("Index2", axis = "index")
print("\n\n",resultframe2)

resultframe3 = dataframe1.drop("Index2", axis = 0)
print("\n\n",resultframe3)
```

### Sorting Data in Dataframes in Pandas

### Handling Duplicates in Pandas

```python
# To find the duplicates, we make use of the duplicated() method.
# It returns True if there is a duplicate and false if there is not

import pandas as pd
data = {
    'Student': ["David", "John", "Jacob", "David", "Steve"],
    'Rank': [1, 4, 3, 5, 2],
    'Marks': [95, 70, 80, 68, 90]
}
custom_index = [f"Index{i}" for i in range(len(data['Student']))]
df = pd.DataFrame(data, index=custom_index)
print(df)

resdf = df.duplicated() # - Gives true when the whole combination is same

rescolspecdf = df.duplicated(subset = ["Student"]) # - gives true based on a specific column

print("\n\nDescribing duplicates through column: \n",rescolspecdf)

print("\n\nDescribing duplicates: \n",resdf)

# Removing duplicates using drop_duplicates() method

res = df.drop_duplicates()

print("\n\n",res)
```

### Grouping data in Pandas

```python
import pandas as pd

data = {
    'Department': ['HR', 'HR', 'IT', 'IT', 'Sales', 'Sales', 'HR'],
    'Employee': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Grace'],
    'Salary': [50000, 55000, 60000, 65000, 45000, 47000, 52000]
}

df = pd.DataFrame(data)
print(df)

# Grouping the data on Department
grpres = df.groupby("Department")

# Printing the first entry
print("\n\nGrouped data: \n", grpres.first())

# Iterating over the groups (display every splitted group)
for name, group in grpres:
  print("\n\n",name)
  print(group)

# Displaying the grouped indexes :
print("\n\n",df.groupby("Department").groups)

# Displaying a specific group :
print("\n\n",df.groupby("Department").get_group("HR"))

# Aggregation operation on groups :
# .sum() , .mean() , .max() , .min() , .count()
```

### Merging data in Pandas

```python
# Merging in pandas is equivalent to SQL joins. inner, left, right, outer
import pandas as pd

# Employee details
df1 = pd.DataFrame({
    'EmployeeID': [1, 2, 3, 4],
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Department': ['HR', 'HR', 'IT', 'IT']
})

# Salary details
df2 = pd.DataFrame({
    'EmployeeID': [1, 2, 3, 5],
    'Salary': [50000, 55000, 60000, 65000]
})

print("DataFrame 1:\n", df1)
print("\nDataFrame 2:\n", df2)

# Inner Merge (Matching Rows)
inner_merge = pd.merge(df1, df2, on='EmployeeID', how='inner')
print("\nInner Merge:\n", inner_merge)

# Left Merge (all rows from left, matching from right)
left_merge = pd.merge(df1, df2, on='EmployeeID', how='left')
print("\nLeft Merge:\n", left_merge)

# Right Merge (all rows from right, matching from left)
right_merge = pd.merge(df1, df2, on='EmployeeID', how='right')
print("\nRight Merge:\n", right_merge)

# Outer merge. All rows from every dataframe
outer_merge = pd.merge(df1, df2, on='EmployeeID', how='outer')
print("\nOuter Merge:\n", outer_merge)
```

### Filtering data in Pandas

```python
import pandas as pd

data = {
    'Department': ['HR', 'HR', 'IT', 'IT', 'Sales', 'Sales', 'HR'],
    'Employee': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Grace'],
    'Salary': [50000, 55000, 60000, 65000, 45000, 47000, 52000]
}

df = pd.DataFrame(data)
print(df)

# Filtering data on salary greater than or equal to 60000
high_salary = df[df['Salary'] >= 60000]
print("\n\n",high_salary)

# Filtering data if department is IT
it_dept = df[df['Department'] == 'IT']
print("\n\n",it_dept)

# Filtering data on HR department and salary greater than 50k
hr_high_salary = df[(df['Department'] == 'HR') & (df['Salary'] > 50000)]
print("\n\n",hr_high_salary)

# Filtering data using isin
hr_sales = df[df['Department'].isin(['HR', 'Sales'])]
print("\n\n",hr_sales)

# Filtering on str.contains() - for matching text
name_with_a = df[df['Employee'].str.contains('a', case=False)]
print("\n\n",name_with_a)
```

### Legend in Matplotlib

Legend is usally a small box at some corner of you graph that describe the frequency of the plotted lines (Basically who's who)

To create or display a legend we use the legend() function that comes within the pyplot submodule



```python
# Adding a legend in your graph

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

arr1 = np.arange(5)
arr2 = np.array([2,4,6,8,10])
arr3 = np.array([5,6,7,8,9])

fig = plt.figure() # This line creates a new blank figure (an empty canvas) where your plot will be drawn. (We do not use above why now? Above we
# only make the one line, so it automatically generated, beside when we have to work with the two line at a time so we have to do it.)
ax = plt.subplot() # Creates a subplot (axes area) inside the figure.(This is made only when multiple line is plotted)

# Plotting two different subplots(line graphs) {label is act as the name}
ax.plot(arr1,arr2, "g--", label = "First_launch")
ax.plot(arr1,arr3, "r:", label = "Second_launch")

#Adding the legend

plt.legend()
plt.show()

#(loc = "upper left", upper right, upper center, lower left, lower right, lower center, center)
```

### Creating Pie Charts

A pie chart is a pictorial representation of data with multiple slices. These slices/portions define the value the amount

To plot pie charts in Matplotlib we make use of pyplot.pie()

This .pie() method has 3 parameters
- the first one is x. It mentions the values
- the second one is labels. String literal mentioning the name of the value
- the third one is autopct. It is a string literal mentioning the percentage a particular portion holds in the whole pie chart.

### Creating a Strip Plot

A strip plot in Seaborn (sns.stripplot()) is a categorical scatter plot that shows individual data points for one categorical variable (x-axis) and one numerical variable (y-axis).

Use Case/ Purpose

| Purpose                           | Description                                                                                        |
| --------------------------------- | -------------------------------------------------------------------------------------------------- |
| **1. Show raw data distribution** | Displays every data point ‚Äî useful when you want to see the actual spread, not just summary stats. |
| **2. Detect outliers**            | Outlier values are easy to spot since all points are plotted.                                      |
| **3. Compare categories**         | Lets you compare numeric values across categorical groups (e.g., total bills across days).         |
| **4. Visualize sample size**      | Dense areas indicate many overlapping data points.                                                 |


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load example dataset
tips = sns.load_dataset("tips")

# Create a basic strip plot
sns.stripplot(data=tips, x="day", y="total_bill", hue = 'sex') #hue, jitter = '0...'

# Add labels and title
plt.xlabel("Day of the Week")
plt.ylabel("Total Bill ($)")
plt.title("Strip Plot of Total Bill by Day")

plt.show()
```

### Step 3: Imputing Missing Values

**1. Mean Imputation (for numerical columns)**

  - Replace missing values with the mean (average) of the column.

  - Works well if data is normally distributed (no major outliers).

*df['body_mass_g'].fillna(df['body_mass_g'].mean(), inplace=True)*

**2. Median Imputation**

  - Replace missing values with the median (middle value) of the column.

  - Better than mean if your data has outliers or skewed distribution.

*df['flipper_length_mm'].fillna(df['flipper_length_mm'].median(), inplace=True)*

**3. Mode Imputation (for categorical columns)**

  - Replace missing values with the most frequent (mode) value.

  - Used for categorical features like sex or species.

*df['sex'].fillna(df['sex'].mode()[0], inplace=True)*

**4. KNN Imputer (More Advanced)**

  - Uses K-Nearest Neighbors algorithm to estimate missing values.

  - Looks for the K most similar rows and fills missing values with the average (or majority) from them.

  - Works well when multiple numerical columns have correlated relationships.

```python
# We are gonna go with mean imputation for this dataset(numerical columns)
# And mode imputation for this dataset(categorical columns)

# Step 1 - Load and inspect missing values

import seaborn as sns
import pandas as pd

# Load dataset
df = sns.load_dataset("penguins")

# See how many missing values exist for the each of the columns
print(df.isnull().sum())
```

#### Step 2 - separate numerical and categorical columns

```python
# Identify numerical and categorical columns
num_cols = df.select_dtypes(include=['float64', 'int64']).columns
cat_cols = df.select_dtypes(include=['object', 'category']).columns
```

```python
# Step 3 - Fill missing values
# Numerical Columns: Fill with Mean

for col in num_cols:
    df[col] = df[col].fillna(df[col].mean())
```

```python
# Categorical Data - Fill with mode

for col in cat_cols:
    mode_values = df[col].mode()
    if not mode_values.empty:
        df[col] = df[col].fillna(mode_values[0])
    else:
        # if the column is completely NaN, fill with a placeholder or keep as NaN
        df[col] = df[col].fillna("Unknown")
```
